## Populando a Tabela dos Continentes (continente)

```python
with open('../data/country-and-continent-codes-list-csv.csv') as file:
    reader = csv.reader(file)
    dicionario = {}
    for row in list(reader)[2:]:
        if not row[1] in dicionario:
            cursor.execute('INSERT INTO continente (id_continente, nome) VALUES (%s, %s)', (row[1], row[0]))
            dicionario[row[1]] = 1
        else:
            pass
```

# Populando tabela de Consumo energético Anual (consumo_anual)

```python
def pop_consumo_anual(fonte_energia):
    fonte_id = ''
    match fonte_energia:
        case 'carvao':
            fonte_id = 0
        case 'gas':
            fonte_id = 1
        case 'hidreletrica':
            fonte_id = 2
        case 'nuclear':
            fonte_id = 3
        case 'petroleo':
            fonte_id = 4
        case 'solar':
            fonte_id = 5

    query = 'INSERT INTO consumo_anual (ano, fk_consumo_fonte_energia, fk_consumo_pais, porcentagem)'
    query += f'VALUES (%s, {fonte_id}, %s, %s)'

    with open(f'../data/energia_{fonte_energia}.csv') as file:
        reader = csv.reader(file)
        _ = next(reader)
        values = []
        for row in reader:
            if (row[1] != '' and len(row[1]) <= 3):
                value = []
                value.append(row[2]) # ano
                value.append(row[1]) # sigla
                value.append(row[3]) # participacao no consumo
                print(value)
                values.append(value)

    cursor.executemany(query, values)
```

# Populando a tabela de Vidas Perdidas em Desastres Naturais por ano (mortes_desastres_naturais)

```python
with open('../data/mortes_desastres.csv') as file:
    reader = csv.reader(file)
    _ = next(reader)

    values = []
    for row in reader:
        if (len(row[1]) <= 3 and row[1] != ""):
            value = []
            value.append(row[1]) # pais
            value.append(row[2]) # ano
            value.append(int(float(row[3]))) # n mortes
            values.append(value)
            print(value)

query = 'INSERT INTO mortes_desastres_naturais (fk_mortes_pais, ano, qntd_vidas)'
query += 'VALUES (%s, %s, %s)'

cursor.executemany(query, values)
```

# Populando a tabela de População anual (populacao_anual)

```python
with open('../data/populacao.csv') as file:
    reader = csv.reader(file)
    anos = next(reader)[4:]
    
    values = []
    for row in reader:
        pais = row[1] # pais
        for i, ano in enumerate(anos):
            valor = row[4 + i]
            print([ano, pais, valor])
            if (ano != '' and pais != '' and valor != '' and pais in teste):
                values.append([ano, int(float(valor)), pais])

query = 'INSERT INTO populacao_anual VALUES (%s, %s, %s)'

cursor.executemany(query, values)
```
